## Transformers.js Inference

Now, let's run inference on the TranslateGemma model using Transformers.js.
First, create a text generation pipeline and then prepare input (list of text messages) to run inference and get the output as desired response.
For reference, you can check the model's page on the Hugging Face model hub under ONNX models section [here](https://huggingface.co/onnx-community/translategemma-text-4b-it-ONNX).

```
%%writefile index.js

// Import the required modules
import { pipeline } from "@huggingface/transformers";

// Create a text generation pipeline
const generator = await pipeline(
  "text-generation",
  "onnx-community/translategemma-text-4b-it-ONNX",
  { dtype: "q4" },
);

// Define the translation input
const messages = [
  {
    role: "user",
    content: [
      {
        type: "text",
        source_lang_code: "en",
        target_lang_code: "fr",
        text: "The weather is beautiful today.",
      },
    ],
  },
];

try {
  // Generate a translation
  const output = await generator(messages, {
    max_new_tokens: 1024,
  });
  console.log(output[0].generated_text.pop().content);
} catch (error) {
  console.error("Error generating translation:", error);
}
```
